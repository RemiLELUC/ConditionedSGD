```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#N = 100
#d = 20


N = 300
d = 100

# load results
ratio_sgd = np.load('ridge_sgd_d{}.npy'.format(str(d)))
ratio_sgd_avg = np.load('ridge_sgd_avg_d{}.npy'.format(str(d)))
ratio_equal = np.load('ridge_equal_d{}.npy'.format(str(d)))
ratio_weighted = np.load('ridge_weighted_d{}.npy'.format(str(d)))
ratio_adafull_avg = np.load('ridge_adafull_avg_d{}.npy'.format(str(d)))
```

```python
tab = np.arange(N+1)
fig,ax = plt.subplots(figsize=(5,5))
# SGD curves
plt.plot(ratio_sgd,
         label='sgd',
         linestyle='dashdot',linewidth=1,marker='s',markevery=0.1,ms=5)

# Polyak-averaging curves
plt.plot(ratio_sgd_avg,
         label='sgd_avg',
         linestyle='dotted',linewidth=1,marker='D',markevery=0.1,ms=4)

# C-SGD equal curves
plt.plot(ratio_equal,
         label=r'csgd($\eta$=0)',
         linestyle='dashed',linewidth=1,marker='^',markevery=0.1,ms=5)

# C-SGD weighted curves
plt.plot(ratio_weighted,
         label=r'csgd($\eta$>0)',
         linestyle='dashed',linewidth=1,marker='o',markevery=0.1,ms=5)

# AdaFullAvg
plt.plot(ratio_adafull_avg,
         label=r'adafull_avg',
         linestyle='dashdot',linewidth=1,marker='*',markevery=0.1,ms=6)
# Graphics and Layout
plt.yscale('log')
#plt.xscale('log')
plt.ylabel(r'Optimaliy Ratio',fontsize=16)
plt.xlabel('Iterations',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(loc='best',fontsize=15)
plt.grid(linestyle='--',which='both',alpha=0.2)
plt.tight_layout()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

plt.show()
```
